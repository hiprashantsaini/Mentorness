# Feature Engineering

#### 1. Feature Extraction

```python
# Extracting new feature using 'CURRENT DATE' and 'DOJ' 
from datetime import datetime
df['DOJ']=pd.to_datetime(df['DOJ'])
df['CURRENT DATE']=pd.to_datetime(df['CURRENT DATE'])
df['WORKING YEARS']=df['CURRENT DATE']-df['DOJ']
```

```python
df['WORKING YEARS']
```

```python
df['WORKING YEARS']=df['WORKING YEARS'].dt.days
```

```python
df['WORKING YEARS']=df['WORKING YEARS']/365.25  # covert in years
```

```python
df
```

```python
sns.boxplot(y=df['WORKING YEARS'])
```

```python
# Remove these outliers 
q1=df['WORKING YEARS'].quantile(0.25)
q3=df['WORKING YEARS'].quantile(0.75)
iqr=q3-q1
minimum=q1-(1.5*iqr)
maximum=q3+(1.5*iqr)
df['WORKING YEARS'].min()
```

```python
print(minimum)
print(maximum)
```

```python
df[(df['WORKING YEARS']<minimum) | (df['WORKING YEARS']> maximum)].value_counts()
```

```python
df['WORKING YEARS']=np.where(
           df['WORKING YEARS']>=maximum,
             maximum,
          df['WORKING YEARS']
)
```

```python
df=df.drop(['DOJ','CURRENT DATE'],axis=1)
df['RATINGS'].min()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['SALARY']),df['SALARY'],test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

### Steps for pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
```

```python
#first step
link1=ColumnTransformer([('encod1',OrdinalEncoder(categories=[['F','M']]),[0]),('encod2',OneHotEncoder(sparse=False,handle_unknown='ignore',drop='first'),[1,3])],remainder='passthrough')
```

```python
link1
```

```python
#Second step
link2=ColumnTransformer([('scale',StandardScaler(),slice(0,15))])
```

```python
link2
```

```python
#Third step
link3=LinearRegression()
# link3=DecisionTreeRegressor()
# link3=RandomForestRegressor()
# link3=GradientBoostingRegressor()
```

### Creating pipeline

```python
pipe=Pipeline([('link1',link1),('link2',link2),('link3',link3)])
```

### Now train pipeline

```python
pipe.fit(X_train,y_train)
```

```python
   pipe.named_steps
```

```python
 pipe.named_steps['link1'].transformers_[0][1]
```

```python
ypred=pipe.predict(X_test)
```

```python
import math
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
```

```python
mean_absolute_error(ypred,y_test)
```

```python
mse=mean_squared_error(ypred,y_test)
mse
```

```python
# Root mean squared error
rmse=math.sqrt(mse)
rmse
```

```python
r2_score(ypred,y_test)
```
